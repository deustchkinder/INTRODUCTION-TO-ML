KNN : Gözlemlerin birbirine olan benzerlikleri üzerinden tahmin yapılır


Bana arkadasını söyle sana kim oldugunu söyleyeyim.

Bir gözlem biriminin kendine en yakın olan  diğer gözlem birimleri hesaplanır, bir gözlem biriminin kendine yakın olan diğer K adet gözlem birimi hesaplanır. Ve bu diğer K adet gözlem biriminin   bağımlı değişkeni ne ise bunun üzerinden ilgili gözlem için tahminde bulunulur.

KNN regresyon ve sınıflandırma problemleri için kullanılabilir





Öklid veya benzeri bir uzaklık yöntemiyle bu gözlem biriminin herbir gözlem birimi arasındaki uzaklığı hesaplayıp, kendisine en yakın olan k tane gözlem biriminin bağımlı değişkenininin ortalamasını almak gerekiyor  

Öklid ya da benzeri bir uzaklık hesabı ile her bir gözleme uzaklık hesaplanır.  Örneğin sonrasında bu gözlem birimine(x1=50,x2=230) yakın olan K = 5 tane gözlem birimi seçilir. (Bizim gözlem birimimiz siyah nokta) Buradaki yakınlık nasıl belirlenir? Örneğin öklid benzeri uzaklık ile belirlenir.
