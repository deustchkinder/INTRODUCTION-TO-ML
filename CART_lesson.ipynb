 **Sınıflandırma Problemleri için Cost/Lost/Objective fonksiyonu**
 
 

Bir sınıflandırma problemi için Gini katsayısını, entropiyi kullanabiliriz. Şimdi buradaki genel mantık şudur: entropi ve gini değerleri aslında bize tahminlerimiizin başarısına yönelik bilgi sağlar.  Elimizde gerçek sınıflar var bu sınıflara ilişkin tahminlerimiz var . Bu gerçek değerler le tahminler arasındaki durumu değerlendirerek bize başarımıza ilişkin bir bilgi verir. Bu basarı metriklerine göre dallanmalara karar verilir.

Bunlara genelde saflık ölcüleri denir. Yani benzerlik benzemezlik olarak düşün. Ne kadar düşük o kadar iyidir. Entropi ve gini değeri bu sekilde sınıflandırma problemleri için hatamızı değerlendirme imkanı sağlar.

**Entropi** : Entropi çeşitlilik demektir. Ne kadar düşük o kadar iyidir. Entropi varyans demektir. Bilgi demektir.Tahmin ettiğimiz değerlere odaklanacak oldugumuzdan dolayı bunların çeşitli olmamasını bekleriz. Bunların düşük olmasını bekleriz.
