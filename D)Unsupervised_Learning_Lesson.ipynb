# Denetimsiz Öğrenme ( Unsupervised Learning )

Karşınıza gözetimsiz öğrenme olarak da çıkabilir. Nedir gözetimsiz öğrenme ?  Önce gözetimli öğrenmenin ne oldugunu hatırlayalım. Elimizdeki bazı feature'lara karşılık bu feature'larda bir şeyler oldugunda ortaya cıktıgı varsayılan bir çıktı, yanıt, bağımlı değişken olduğunda buna denetimli öğrenme problemleri diyorduk.

Eğer ilgilendiğimiz problemde bir bağımlı değişken , hedef değişken yok ise yani ilgili gözlem birimlerinde meydana gelen gözlemlere karşılık ortaya ne cıktıgı bilgisi yoksa diğer ifadesiyle bir label yoksa etiket yoksa bunlar da gözetimsiz öğrenme problemleridir. 

Birlikte değerlendireceğimiz 3 tane yöntem var.
 - K-Ortalamalar(K-Means)
 - Hiyerarsik Kümeleme Analizi 
 - Temel Bileşen Analizi (PCA)

## K-Ortalamalar ( K-Means )

Amaç, gözlemleri birbirlerine olan benzerliklerine göre kümelere ayırmaktır.



Buradaki bazı gözlemler birbirlerine olan yakınlıklarına göre çeşitli uzaklık metrikleri kullanılarak kümelere ayrılmış . K-Means deki temel amacımız, gözlem birimlerini kümelere ayırmaktır.

DİKKAT!! Sanki sınıflandırma yöntemlerinde de sınıflara ayırıyorduk. Evet **sınıflara** ayırıyorduk ve sınıflar belliydi çalışmanın başında.  Dolayısıyla çalışmanın başında belli olan sınıflara göre bazı ayırt edici noktaları karar kurallarını, fonksiyonlarını belirliyorduk. Bu belirlemelerden sonra bir gözlem birimi geldiğinde onun hangi sınıfa ait oldugunu tahmin ediyorduk. Fakat kümeleme işleminde elimizde böyle bir bağımlı değişken yada bir sınıf bilgisi yok. Dolayısıyla gözlem birimlerini kendi içinde birbirlerine benzerliklerine göre gruplara ayırıyoruz.Öbekliyoruz. Kümeliyoruz.

Peki nasıl çalışır ??

Adım 1 : Küme sayısı belirlenir.

Adım 2 : Rastgele k adet merkez seçilir.

Adım 3 : Her gözlem için k merkezlere olan uzaklıklar hesaplanır.

Adım 4 : Her gözlem en yakın olduğu merkeze yani kümeye atanır.

Adım 5 : Atama işlemlerinden sonra oluşan kümeler için tekrar merkez hesaplamaları yapılır. 

Adım 6 : Bu işlem belirlenen bir iterasyon adedince tekrar edilir ve küme içi hata kareler toplamlarının toplamının (total within-cluster variation) minimum olduğu durumdaki gözlemlerin kümelenme yapısı nihai kümelenme olarak seçilir.


Bir şekilde en optimum küme merkezine gitmeye çalışıyoruz. İyi ama nasıl ?? Şimdi elimizde bazı kümeler var , merkezleri var bunların. Ve gözlem birimleri var etrafında. Şimdi öyle bir şey yapmam lazım ki bu kümeler kendi içinde homojen, birbirlerine göre heterojen olsun. E bunu nasıl yapabilirim ? Bunu matematiksel bir şekilde nasıl belirleyebilirim sorusunun yanıtı bazı metrikler ile ortaya cıkarıyoruz. SSE olarak yada SSR olarak karsınıza gelebilir. Yani bu ifadedir => total within-cluster variation

Şimdi bu ne demek ? 

Bir merkez oldugunu düşünelim bu bir gözlem birimi. Bu merkezdeki gözlem biriminin değerleri ile etrafındaki diğer gözlem birimlerinin değerlerinin farklarının karelerini alıp topladıgımızda bu kümenin kendi içindeki hata karesini buluruz. 

Özetle problemimiz şuydu bu merkezleri nasıl belirlicem. Küme içi birbirlerine benzerlik çok olmalı , homojen olmalı, birbirlerine benzer olmalı gözlemler . Kümeler arası heterojen olmalı. 

SSE en düşük oluncaya kadar devam edilir, merkezler belirlenmesi için. 

Amacımız, kümeler içindeki homojenliği cok yüksek yapmak diğer ifadesiyle benzerlik demek farkları kücük olmak demek.Birbirlerine yakın gözlemler olmak demek.Dolayısıyla ne kadar kücük SSE verirse bir merkez artık o merkezin örneğin 5 kümeleme yapılacaksa nihai merkez oldugu bu 4 5 kümenin belirlenir.


Özetle, K-means iteratif bir şekilde önce rastgele atamış oldugu merkezleri  değiştirerek merkezin nerede olması gerektiğine karar verir. Merkezin nerede olması gerektiğini matematiksel bir şekilde  SSE diğer ifadesiyle SSR değerlerine bakarak değerlendirir ve bu sekilde seçer. 

Peki bir başka problem var. Küme sayısını 4 dedik 5 dedik bunu böyle kafamızdan mı vericez.  DİKKAT! k 'nın kaç olacağının belirlenmesi bu yöntemin dezavantajlarından biridir!!! Bunun matematiksel bir yolu yok mu ? küme sayısı belirlemek için ? Evet var. En optimum küme sayısının ne olması gerektiğini göreceğiz.

#   Hiyerarşik Kümeleme Analizi ( Hierarchical Cluster Analysis )

Kümeleme kavramını biliyoruz. Gözlem birimleri birbirlerine olan benzerliklerine göre kümelere ayrılıyordu. Burada da yine kümelere ayırma işlemi var. Fakat bu kümelere ayırma işlemi hiyerarşik bir şekilde gerçekleşiyor. Nasıl yani?





Birleştirici ve bölümleyici olarak 2 genel başlığımız var.  Birleştirici yöntemde çalışmanın başında bütün gözlem birimleri tek başına bir küme gibi düşünülüp, yukaıya doğru iki gözlem birimi bir araya gelip bir küme oluşturur ya da 3 gözlem birimi bir araya gelip bir küme oluşturur gibi yukarıya doğru birleştirerek işlemler yapılır. Bölümleyici kümeleme de ise  tüm gözlemler bir arada oldugunda bu tek bir kümedir.Ondan sonra iki küme elde edilir.Diye diye bölünerek bölümleyici bir şekilde aşağıya doğru tüm veri seti herbir gözlem tek basına kalacak sekilde aşağıya doğru oluşturulur.

Hiyerarşik Kümeleme yöntemi aşağıya doğru yada yukarıya doğru şekilde gözlem birimlerini kümelere sokmaya dayanmaktadır.

Amaç, gözlemleri birbirlerine olan benzerliklerine göre alt kümelere ayırmaktır.


Bunun k-means den ne farkı var ?? Ne getirir , ne götürür nasıl yaklaşmak lazım?

K-means de küme oluşturma sürecine dışarıdan müdahale edemiyorduk. Dolayısıyla bir gözlemleme imkanımız yoktu. Ama burada şöyle bir şansımız var . Hiyerarşik kümelemenin avantajı bize belirli noktalardan şu şekilde yatay çizgiler çizerek görsel teknik üzerinden çeşitli kümelenme seviyelerinde yeni kümelenmeri tanımlayabiliyor olmasıdır. 


# Temel Bileşen Analizi ( Principal Component Analysis )

Temel fikir, çok değişkenli verinin ana özelliklerini daha az sayıda değişken/bileşen ile temsil etmektir.

Diğer bir ifade ile: kücük miktarda bir bilgi kaybını göze alıp değişken boyutunu azaltmaktır.

Boyut indirgeme kaygımız neden var ? Örneğin doğrusal regresyon problemlerinde  çoklu doğrusal bağlantı probleminden kurtulmak istiyor olabiliriz. Örneğin bir yüz tanıma probleminde resimlere filtre yapma ihtiyacı hissedebiliriz gibi sebeplerle boyut indirgeme ihtiyacımız oldugunda bu yöntemi kullanırız.

Boyut indirgeme yaklaşımıdır. Veri setinin boyutunu kücük miktarda bilgi kaybını göze alarak indirgeme işlemidir.

Görüntü işleme problemlerinde, regresyon problemlerinde yani çok boyutluluk söz konusu oldugunda kullanılabilir.
Çok boyutlu veriyi görselleştirmek için de kullanılır.

Temel Bileşen Analizi bir araçtır. Birçok farklı amaç ile kullanılıyor olabilir.

**DİKKAT !**

Temel Bileşen Analizi  veri setini bağımsız değişkenlerin doğrusal kombinasyonları ile ifade edilen bileşenleri indirger.Dolayısıyla bu bileşenler arasında korelasyon yoktur. Bu bileşenler korelasyonsuzdur. 

Bu bileşenleri teknik olarak nasıl cıkarır?
Değişken gruplarının varyanslarını ifade eden özdeğerler ve veri setindeki değişkenleri gruplandırır.Yani gruplar arasında en fazla varyansa sahip olan gruplar en önemli gruplardır ki işte bunlar asal bileşenlerdir.
